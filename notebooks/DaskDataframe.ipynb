```python
mini_index = pd.date_range(name="time",start='2018-04-24', end='2018-04-25', freq='10s')
norm_index = pd.date_range(name="time",start='2018-04-24', end='2018-04-25', freq='5min')
```

```python
def gen_df(index, interval):
    vals = []
    for _ in index:
        v = np.random.uniform(low=0, high=5, size=len(feat))
        dct = dict(zip(feat, v))
        dct.update({"interval": interval})
        vals.append(dct)
        
    return pd.DataFrame(vals, index=index)
```

```python
df_mini = gen_df(mini_index, interval="10s")
```

```python
df_norm = gen_df(norm_index, interval="5min")
```

```python
df_mini.reset_index(inplace=True)
df_norm.reset_index(inplace=True)
```
